## Bode and Polar Plots

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style

style.use("classic")

%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

#### First order system

$$G(s) = \frac{1}{\tau s + 1}$$

#### Second order system

$$G(s) = \frac{1}{\tau^2 s^2 + 2 \tau \xi s + 1}$$

where $\xi$ is the dampening constant.

* $\xi < 1$ : Under-damped system
* $\xi = 1$ : Critically damped system
* $\xi > 1$ : Over-damped system

```python
tau = 1.33
omega = np.linspace(0.01, 100, 100000)
```

### Bode plots

#### First order system

```python
AR_by_Kp = 1 / np.sqrt(1 + (omega * tau) ** 2)
phi = np.arctan(-omega * tau)


plt.figure(facecolor='white')
plt.grid()
plt.loglog(omega, AR_by_Kp, color='black', linewidth=1.2)
plt.xlabel(r"$\omega$", fontsize=16)
plt.ylabel(r"$\frac{A.R.}{K_p}$", fontsize=16)

plt.figure(facecolor='white')
plt.grid()
plt.semilogx(omega, phi, color='black', linewidth=1.2)
plt.xlabel(r"$\omega$", fontsize=16)
plt.ylabel(r"$\phi = \arctan(-\omega \tau)$", fontsize=16);
```

#### Second order system

```python
def SecOrderAR_by_Kp(xi):
    return 1 / np.sqrt((1 - (omega * tau) ** 2) ** 2 + (2 * omega * tau * xi) ** 2)


def SecOrderPhi(xi):
    return -np.pi / 2 - np.arctan(-(1 - (omega * tau) ** 2) / (2 * omega * tau * xi))


plt.figure(facecolor='white')
plt.grid()
plt.loglog(omega, SecOrderAR_by_Kp(0.10), color='r', linewidth=1.2, label=r'$\xi = 0.1$')
plt.loglog(omega, SecOrderAR_by_Kp(0.50), color='g', linewidth=1.2, label=r'$\xi = 0.5$')
plt.loglog(omega, SecOrderAR_by_Kp(1.00), color='b', linewidth=1.2, label=r'$\xi = 1$')
plt.loglog(omega, SecOrderAR_by_Kp(2.00), color='k', linewidth=1.2, label=r'$\xi = 2$')
plt.xlabel(r"$\omega$", fontsize=16)
plt.ylabel(r"$\frac{A.R.}{K_p} = \frac{1}{\sqrt{(1 - \omega^2 \tau^2)^2 + (2 \omega \tau \xi)^2}}$", fontsize=16)
plt.legend(loc='best', fontsize=8)

plt.figure(facecolor='white')
plt.grid()
plt.semilogx(omega, 180 * SecOrderPhi(0.10) / np.pi, color='r', linewidth=1.2, label=r'$\xi = 0.1$')
plt.semilogx(omega, 180 * SecOrderPhi(0.50) / np.pi, color='g', linewidth=1.2, label=r'$\xi = 0.5$')
plt.semilogx(omega, 180 * SecOrderPhi(1.00) / np.pi, color='b', linewidth=1.2, label=r'$\xi = 1$')
plt.semilogx(omega, 180 * SecOrderPhi(2.00) / np.pi, color='k', linewidth=1.2, label=r'$\xi = 2$')
plt.xlabel(r"$\omega$", fontsize=16)
plt.ylabel(r"$\phi= \arctan(\frac{-2 \omega \tau \xi}{1 - \omega^2 \tau^2})$", fontsize=16)
plt.legend(loc='best', fontsize=8);
```

### Polar plots

#### First order system

$$G(j \omega) = \frac{1}{j \tau \omega + 1} = re^{i\phi}$$

```python
plt.figure(facecolor='white', figsize=(7, 7))
plt.polar(phi, AR_by_Kp, 'b');
```

#### Second order system

```python
plt.figure(facecolor='white', figsize=(9, 7))
plt.polar(SecOrderPhi(0.5), SecOrderAR_by_Kp(0.5), 'r', label=r'$\xi=0.5$')
plt.polar(SecOrderPhi(1.0), SecOrderAR_by_Kp(1.0), 'g', label=r'$\xi=1.0$')
plt.polar(SecOrderPhi(1.5), SecOrderAR_by_Kp(1.5), 'b', label=r'$\xi=1.5$')
plt.polar(SecOrderPhi(2.0), SecOrderAR_by_Kp(2.0), 'k', label=r'$\xi=2.0$')
plt.legend(loc='best', bbox_to_anchor=[1.15, 1], fontsize=8);
```

#### First order sytem  with dead time

$$G(s) = \frac{1}{\tau s + 1} - \frac{K e^{-\tau_d s}}{\tau s + 1}$$
where $\tau_d$ is the dead time constant.

```python
K = 1
taud = 1

phi1phi2 = -omega * taud + phi

plt.figure(figsize=(9, 7))
plt.polar(phi1phi2 , K * AR_by_Kp, 'b');
```

#### Second order sytem  with dead time

```python
plt.figure(figsize=(9, 7))
plt.polar(SecOrderPhi(0.5) - omega * taud, K * SecOrderAR_by_Kp(0.5), 'r', label=r'$\xi=0.5$')
plt.polar(SecOrderPhi(1.0) - omega * taud, K * SecOrderAR_by_Kp(1.0), 'g', label=r'$\xi=1.0$')
plt.polar(SecOrderPhi(1.5) - omega * taud, K * SecOrderAR_by_Kp(1.5), 'b', label=r'$\xi=1.5$')
plt.polar(SecOrderPhi(2.0) - omega * taud, K * SecOrderAR_by_Kp(2.0), 'k', label=r'$\xi=2.0$')
plt.legend(loc='best', bbox_to_anchor=[1.15, 1], fontsize=8);
```
